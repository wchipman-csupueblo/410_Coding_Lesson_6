```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('ramen-ratings.csv')
data.head()
```

```python
sns.barplot(data=data, x='Style', y='Stars')
```

```python
sns.catplot(data=data, x='Style', y='Stars', kind='bar')
```

```python
g = sns.catplot(data=data, x='Style', y='Stars', kind='bar')
for ax in g.axes.flat:
    ax.set(title='Rating by Ramen Type', xlabel='Ramen Type')
```

```python
g = sns.catplot(data=data.query('Country in ["Japan", "India", "Taiwan", "USA"]'),x='Style', y='Stars', kind='bar', col='Country', col_wrap=2)
```

```python
g = sns.catplot(data=data.query('Country in ["Japan", "India", "Taiwan", "USA"]'), x='Style', y='Stars', kind='bar', col='Country', col_wrap=2)
countries = data.query('Country in ["Japan", "India", "Taiwan", "USA"]').Country.unique()
for i, ax in enumerate(g.axes.flat):
    ax.set(title=countries[i] + ' Ratings by Ramen Type', xlabel='Ramen Type')
```
